# Bayesian Filtering: Kalman Filters Are Bayesian

In the first set of notes, we formulated a linear Kalman Filter based on the premise of averaging prior beliefs with new information. In this section, we'll prove that a Kalman filter is actually just a specific instance of a Bayesian filter. 

## Assumptions

1. we interpret the Kalman filter's current *posterior* estimate $\hat{x}_k$ and corresponding covariance $P$ as the parameters of a Gaussian belief.

$$
X_{k - 1} \mid z_{k - 1} \sim \mathcal{N}(\hat{x}_{k-1}, P_{k-1}) 
$$

2. The Kalman Filter's prediction model is a linear transform plus some error $W$, whose mean is zero with covariance $Q$:

$$
X_k = FX_{k - 1} + W, ~~~~ W \sim \mathcal{N}(0, Q)
$$

## Step 1: Prediction Step Proof

We said that the prediction step of the Kalman Filter was

$$
P_k^- = F P^+_{k - 1} F^T + Q 
$$

and that 

$$
\hat{x}_k^- = F\hat{x}_{k - 1}^+
$$

Where the (-/+) superscripts indicate before and after the belief update of the corresponding timestep $k$. 

Our assertion is that 

$$
p(x_k \mid z_{k - 1}) = \mathcal{N}(\hat{x}_k, F P^+_{k - 1} F^T + Q)
$$

In other words, the predictions of state $x^-_k$ and uncertainty $P^-_k$ produced by a Kalman Filter *are also* the mean and covariance resulting from a Bayesian prediction step. 




### Step 1a: Proving the Mean

By definition, expectation is a linear operation. Therefore we find

$$
\begin{align}
\mathbb{E}[X_k] &= \mathbb{E}[FX_{k - 1} + W] \\
&=F\mathbb{E}[X_{k - 1}] + \mathbb{E}[W] \\
&=F\mathbb{E}[X_{k - 1}] \\
&=F\hat{x}_{k-1}
\end{align}
$$

Because we said that $W$ has zero mean, that term disappears. 

This completes proving the first part of the proof:

$$
\boxed{ \mathbb{E}[X_k] = F\hat{x}_{k-1}}
$$

### Step 1b: Proving the Covariance

By the definition of covariance

$$
P_{k - 1} = \mathrm{Cov}(X_{k - 1}) = \mathbb{E}[(X_{k-1} - \hat{x}_{k - 1})(X_{k-1} - \hat{x}_{k - 1})^T] 
$$

And also by definition

$$
P_{k} = \mathrm{Cov}(X_k) = \mathbb{E}[(X_k - \hat{x}_k)(X_k - \hat{x}_k)^T]
$$

Substituing $X_k = FX_{k - 1} + W$ followed by some factoring of $F$ we see

$$
\begin{align}
P_{k} &= \mathbb{E}[(FX_{k - 1} + W - F\hat{x}_{k - 1})(FX_{k - 1} + W - F\hat{x}_{k - 1}))^T] \\
&= \mathbb{E}[(F(X_{k - 1} - \hat{x}_{k - 1}) + W)(F(X_{k - 1} - \hat{x}_{k - 1}) + W)^T]
\end{align}
$$

Now, recall (or look up) the fact that

$$
(A + B)(A + B)^T = AA^T + AB^T + BA^T + BB^T
$$

And then let

\begin{align}
A &= F(X_{k - 1} - \hat{x}_{k - 1}) \\
B &= W
\end{align}

So we have 

\begin{align}
P_k &= \mathbb{E}[AA^T + AB^T + BA^T + BB^T] \\
&= \mathbb{E}[AA^T] + \mathbb{E}[AB^T] + \mathbb{E}[BA^T] + \mathbb{E}[BB^T]
\end{align}

We will address each one of these terms individually. 

---
Looking at the first term in this sum, and considering that $F$ is a constant, we get

\begin{align}
\mathbb{E}[AA^T] &= \mathbb{E}[(F(X_{k - 1} - \hat{x}_{k - 1}))(F(X_{k - 1} - \hat{x}_{k - 1}))^T] \\
&= \mathbb{E}[F(X_{k - 1} - \hat{x}_{k - 1})(X_{k - 1} - \hat{x}_{k - 1})^TF^T] \\
&= F\mathbb{E}[(X_{k - 1} - \hat{x}_{k - 1})(X_{k - 1} - \hat{x}_{k - 1})^T]F^T
\end{align} 

Then, we remember

$$P_{k - 1} = \mathbb{E}[(X_{k-1} - \hat{x}_{k - 1})(X_{k-1} - \hat{x}_{k - 1})^T] $$

and substitute to get

$$
\mathbb{E}[AA^T] = FP_{k - 1}F^T
$$

---

The next term is $\mathbb{E}[AB^T]$. For this term we recall that the process noise $W$ is independent of $X_{k-1}$, and that for independent random variables $X$ and $Y$

$$
\mathbb{E}[XY^T] = \mathbb{E}[X]\mathbb{E}[Y]^T
$$

Therefore,

\begin{align}
\mathbb{E}[AB^T] &= \mathbb{E}[(F(X_{k-1} - \hat{x}_{k - 1})W^T] \\
&= \mathbb{E}[F(X_{k-1} - \hat{x}_{k - 1})]\mathbb{E}[W]
\end{align}

But we said that $W$ has zero

$$\mathbb{E}[W] = 0$$

And therefore $\mathbb{E}[AB^T] = 0$. Likewise, the second term $\mathbb{E}[BA^T]$ also goes away.

---

The final term to consider is $\mathbb{E}[BB^T] = WW^T$

So now we have 

$$
P_{k} = FP_{k-1}F^T + \mathbb{E}[WW^T]
$$

And we recall that
1. The mean of $W$ is $\mu_W = 0$.
2. Q is the covariance

$$Q = \mathbb{E}[(W - 0)(W - 0)^T] = \mathbb{E}[WW^T]$$

And we have completed the proof of

$$
\boxed{ P_{k} = FP_{k-1}F^T + Q }
$$

and

$$
\boxed{ \mathbb{E}[X_k] = F\hat{x}_{k-1}}
$$
