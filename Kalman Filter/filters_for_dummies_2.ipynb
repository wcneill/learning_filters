In this section, we assume that all quantities are scalar, and derive the formula for updating our estimations and their uncertainties when we receive new information. 

Also, when we say "uncertainty", we are talking specifically about statistical variance and will use the corresponding common notation for variance: $\sigma^2$

## Weighted Average of Information: Old and New

Let's say we estimate our position on a line in one dimension as $x_{\mathrm{est}}$ and we assign some uncertainty to that estimation as $\sigma_{\mathrm{est}}^2$. Then, we recieve a measurement of our position $x_{\mathrm{meas}}$ that also has some uncertainty $\sigma_{\mathrm{meas}}^2$, perhaps due to the limitations of the sensor. Now we have two estimates of position

$$
\begin{align}
e_1 &= x_{\mathrm{est}} \pm \sigma_{\mathrm{est}}\\ 
e_2 &= x_{\mathrm{meas}} \pm \sigma_{\mathrm{meas}}
\end{align}
$$

How do we resolve the difference in these two values? One method that might come to mind is to come up with a new estimate which is the weighted average of the two. 

This leads naturally to the question of how to weight each term in the average. One method would be to incorporate the uncertainty into the weighting. As uncertainty goes up, that estimate should have less weight in the average. If uncertainty decreases, we'd like the corresponding estimate to weigh more.  

We could start with the proportion of uncertainty associated with an estimate to the total uncertainty of the estimate and the measurement:


<br>
$$
\begin{align}
w_{\mathrm{est}} &= \frac{\sigma_{\mathrm{est}}^2}{\sigma_{\mathrm{meas}}^2+\sigma_{\mathrm{est}}^2} \\\\
w_{\mathrm{meas}} &= \frac{\sigma_{\mathrm{meas}}^2}{\sigma_{\mathrm{meas}}^2+\sigma_{\mathrm{est}}^2}
\end{align}
$$
<br>

But there's something wrong with this... as uncertainty decreases, the weights would decrease... That's not what we want. We want to give more weight to measurements or estimates that we are more certain about. But, we can keep this idea of proportions with a slight modification. What if we instead made the weight

<br>
$$
\begin{align}
w_{\mathrm{est}} &= 1 - \frac{\sigma_{\mathrm{est}}^2}{\sigma_{\mathrm{meas}}^2+\sigma_{\mathrm{est}}^2} \\\\
w_{\mathrm{meas}} &= 1 - \frac{\sigma_{\mathrm{meas}}^2}{\sigma_{\mathrm{meas}}^2+\sigma_{\mathrm{est}}^2}
\end{align}
$$
<br>

Now, as uncertainty decreases towards zero, the weight associated with that uncertainty increases towards 1. And, since we are using the proportion of individual estimate uncertainty to total uncertainty, we know that all the uncertainties must sum to 1. 

<br>
$$
\begin{align}
& w_{\mathrm{est}} + w_{\mathrm{meas}} = 1  \\
\Rightarrow & w_{\mathrm{est}} = 1 - w_{\mathrm{meas}}
\end{align}
$$
<br>

This is convenient because it allows us to simplify the weight formulas through a simple substitution:

<br>
$$
\begin{align}
w_{\mathrm{est}} = 1 - w_{\mathrm{meas}} = \frac{\sigma_{\mathrm{meas}}^2}{\sigma_{\mathrm{meas}}^2+\sigma_{\mathrm{est}}^2} \\\\
w_{\mathrm{meas}} = 1 - w_{\mathrm{est}} = \frac{\sigma_{\mathrm{est}}^2}{\sigma_{\mathrm{meas}}^2+\sigma_{\mathrm{est}}^2} \\\\
\end{align}
$$
<br>

Now that we have defined how to calculate the weights associated with each uncertainty, we can go ahead and find the weighted average of the estimate and measurement. 

<br>
$$
x_{\mathrm{new}} = w_{\mathrm{est}} x_{\mathrm{est}} + w_{\mathrm{meas}} x_{\mathrm{meas}}
$$
<br>

### Aligning the Formula With Intuition

We could stop here, but to align our formulas with our intuition, we need to take one more step. Recall that we first estimated our position, then recieved a measurement from a sensor, and we noticed there was a ***difference*** between them, and now want to update our estimate. It would be nice, then, if our update formula could match that thought process. 

By recalling that $$w_{\mathrm{est}} = 1 - w_{\mathrm{meas}}$$ we can do a small bit of algebra (ugh):

<br>
$$
\begin{align}
x_{\mathrm{new}} &= (1 - w_{\mathrm{meas}}) x_{\mathrm{est}} + w_{\mathrm{meas}} x_{\mathrm{meas}} \\
&= x_{\mathrm{est}} - w_{\mathrm{meas}} x_{\mathrm{est}} + w_{\mathrm{meas}} x_{\mathrm{meas}} \\
&= x_{\mathrm{est}} + w_{\mathrm{meas}} \left(x_{\mathrm{meas}} - x_{\mathrm{est}} \right)
\end{align}
$$
<br>

Here we see the update in a new light. Namely we are updating our original estimate through adding a corrective term that is the weighted difference between the estimate and the measurement. The correction is scaled by the measurement weight, which increases as the measurement becomes more trustworthy and decreases as the measurement becomes more uncertain.

### Filter Notation

Here we can shift start to shift towards standard Kalman filter notation by re-writing  $\sigma_{\mathrm{est}}^2 = P$ and $\sigma_{\mathrm{meas}}^2 = R$.

We will then re-label the measurement weight as

$$
\begin{align}
\boxed{K  = \frac{P}{P + R} = \frac{\sigma_{\mathrm{est}}^2}{\sigma_{\mathrm{est}}^2+\sigma_{\mathrm{meas}}^2}}
\end{align}
$$

And we can rewrite our update equation accordingly:

$$
\begin{align}
\boxed{x_{\mathrm{new}} = x_{\mathrm{est}} + K \left(x_{\mathrm{meas}} - x_{\mathrm{est}} \right)}
\end{align}
$$

This gives us the notation commonly used in Kalman filter literature.  

## Updating Variance

The formula we derived in the previous section shows how we update our estimate based on new information. This new estimate will still come with a measure of uncertainty, so we need to find out what that uncertainty is. 

<br>
$$
P_{\mathrm{new}} = (1 - K)P_{\mathrm{old}}
$$
<br>

With some inspection, we see that as the certainty associated with the measurement increases, $K \rightarrow 1$, and there fore $P_{\mathrm{new}} \rightarrow 0$. This makes sense, because if we recieve a measurement that we are 100 percent certain of, our uncertainty should go to zero. We discard the old estimate completely, and just take the value of the new information as complete truth (Intuitive, but also a consequence of the weighted average we defined above).

Likewise, as the uncertainty associated with the measurement increases, $K \rightarrow 0$ and $P_{\mathrm{new}} \rightarrow P_{\mathrm{old}}$. This, too, makes sense. Because if we have zero confidence in the new information, we would just discard it and keep the old estimate. This too is both intuitive and a direct consequence of the weighted average we defined above. 

### Where Does $x_{\mathrm{est}}$ Come From, Anyway?

This is also where the algorithm we are working towards begins to show itself.

Our position estimate $x_{\mathrm{est}}$ is the output of previous step in the algorithm we are developing, which predicts our next position with an *imperfect* mathematical model of the system. The general form of such a model is 

<br>
$$x_{i+1} = f(x_{i})$$
<br>

### Okay, where does $P_{\mathrm{old}}$ Come From, Then?

***This is where we are going to kick things up from 1-dimensional scalars to start using linear algebra. This is because we need to start generalizing the algorithm so that it works for problems in greater than 1 dimension. We will fully generalize the equations later, though.*

We now know how $x_{\mathrm{est}} = x_{\mathrm{i + 1}}$ is produced, so the next question is "How did we get $P_{i + 1}$?". 

To answer that question, we will step away from our specific example for a moment. We will define a random variable $x$, and with it we will define a second random variable as a $x$ scaled by some constant factor $a$. This is just a linear transformation, and we can think about $y$ as our model's prediction, and $x$ as the input to the model. 

$$
y = ax
$$

Here, $x$ is our input, and has some uncertainty $\sigma^2_x = Var(x)$ associated with it. We would like to know $\sigma^2_y = Var(y)$. It's pretty easy to derive:

<br>
$$
\begin{align}
Var(y) &= Var(ax) \\
&= \mathrm{E}[(ax - a\mu_x)^2] \\
&= \mathrm{E}[(a(x - \mu_x))^2]\\
&= \mathrm{E}[a^2(x - \mu_x)^2]\\
&= a^2\mathrm{E}[(x - \mu_x)^2]\\
&= a^2Var(x)
\end{align}
$$
<br>

So we see that the new uncertainty is simply the old uncertainty multiplied by $a^2$. Pretty straight forward. 

Now, let's kick this up to a higher dimension. Recalling that in linear algebra $AA^T$ is the higher dimension analogue to $a^2$ in one dimension, we get the formula for covariance, $\Sigma$, of scaled random variables:

<br>
$$
\begin{align}
\Sigma_y &= A\Sigma_x A^T 
\end{align}
$$
<br>

Moving back to our predictive model, what is the "scaling factor" $A$?. We might recall that the derivative of a 1-D function tells us how the output scales up or down with respect to its input. We might then recall that the Jacobian matrix $J$ is the multi-dimensional analogue to the derivative. This chain of logic naturally leads us to

<br>
$$
\begin{align}
 J \Sigma J^T 
\end{align}
$$
<br>

It's worth re-iterating that the Jacobian can be thought of as a measure of the sensitivity of a function's output to changes in the input near some point. "Rise over Run" as the slope of a line is forever engrained in my mind from grade school. This is nothing but a fancy, higher dimensional version of that exact concept. We can measure the amount of change $y$ experiences ($\delta y$) with respect to a change in $x$, ($\delta x$), as:

<br>
$$
\delta\mathbf{y} = J(\mathbf{x}_0)\delta\mathbf{x}
$$
<br>

So, if we take this knowledge back to our example, $P_{\mathrm{i}}$ is simply transformed (scaled, sheared, rotated) by the sensitivity of the system at the point being evaluated:

<br>
$$
P_{\mathrm{i}} \approx J P_{i-1} J^T 
$$
<br>

But, that only gives us the affects of the uncertainty of the inputs on the output. There is also uncertainty in our predictive model itself. Recall that "All models are wrong, but some are useful"... We then add a term $Q$, which represents the intrinsic uncertainty in the model: 

<br>
$$
\boxed{P_{\mathrm{i}} = J P_{i-1} J^T + Q}
$$
<br>

## How About That Algorithm?

So, we have derived some formulas that we can use to update our belief or estimation of position based on new information. But, what we are working towards isn't just a set of formulas, but an algorithm which is the Kalman filter. 

The purpose of a Kalman filter is to combine all available information about a system to produce the best estimate of the system state and the uncertainty associated with that estimate. In our simple case of moving a long a 1-D line, the state is just position. The information we are fusing together to achieve the best possible estimate is the sensor measurement and some prior estimation... but we haven't talked about where that prior information came from. 

At this point, I want to explicitly outline that algorithm. You will see that what we have derived thus far is part of that larger algorithm. 

<div style="text-align: center;">
  <pre style="display: inline-block; text-align: left; background: transparent; border: none; padding: 0; margin: 0; font-family: monospace;">
Previous Estimate
        │
        ▼
 ┌────────────┐
 │ Prediction │
 └────────────┘
        │
        ▼
 Predicted State
        │
 Measurement Arrives
        │
        ▼
 ┌────────────────────┐
 │ Measurement Update │
 └────────────────────┘
        │
        ▼
 Updated Estimate
  </pre>
</div>

We can see how the math we discussed follows the algorithm above:

1. Our initial position estimate $x_{\mathrm{est}}$ can be viewed as the output of the first step in the algorithm which is a prediction. This prediction came from some *imperfect* model of the system, that takes the current state as input and predicts next state. It had some associated uncertainty that we called $P_{\mathrm{old}}$. The general form of such a model is $$x_{\mathrm{i+1}} = f(x_{\mathrm{i}})$$   
2. Next, we recieved a measurement $x_{\mathrm{meas}}$, with some uncertainty $R$. 
3. Then, we calculated an updated estimate $x_{\mathrm{new}}$ using a weighted average of our estimate and the measurement. 
4. Finally, we calculated an updated measure of uncertainty $P_{\mathrm{new}}$ associated with $x_{\mathrm{new}}$.

## Summary so Far

We built intuition abou an algorithm in which we successively estimate our current state and then update that estimate with new information. We hope to achieve two important results:

- A better estimate.
- Decreased uncertainty of our estimate.

So the output of our algorithm at each step is always two things:
1. An estimate of our state.
2. A measure of the uncertainty of that estimate. 

### Step 1:
Our algorithm begins with some initial state $x_i^+$, like position, perhaps obtained from a measurement. This initial state comes with some amount of uncertainty $P_i^+$. 

### Step 2:
To get a prediction of what our state will be at the next time step, we use a predictive model of the form $$x_{i + 1}^- = f(x_i^+)$$ where $x_i^-$ indicates an *initial* estimate at timestep $i$ and $x_i^+$ indicates an *updated* estimate at timestep $i$. 

### Step 3: 
We calculate the amount of uncertainty in the input $P_i^+$, that propagates through the model to the output's uncertainty $P_{i + 1}^-$. 

$$
P_{i+1}^- = J P_{i}^+ J^T + Q
$$

Where $J$ is the Jacobian of the model evaluated at $x_i^+$. 


### Step 4:

We receive new information $z_{i + 1}$ with uncertainty $R$ at time $t_{i + 1}$. We update our old estimate with the new information: 


Where $K$ is the proportion of total uncertainty attributed to the estimate:

$$
\begin{align}
K  = \frac{P}{P + R} = \frac{\sigma_{\mathrm{est}}^2}{\sigma_{\mathrm{est}}^2+\sigma_{\mathrm{meas}}^2}
\end{align}
$$

### Step 5:

We calculate the uncertainty of this update with (in the 1-D case) as 

$$
P_{i + 1}^+ = (1 - K)P_{i + 1}^-
$$

### Step 6:

Repeat. 

### A Note on Dimensionality

Notice that we started to talk about higher dimensions for the input uncertainty propogation. We used covariance matrices and Jacobians to propagate uncertainty of our input to see how it affects the uncertainty of the output.

However, our measurement update equations still contain two simplifying assumptions:

1. The measurement is a direct observation of the state being estimated.
2. The Kalman gain \(K\) is treated as a scalar weight.

We'll remove those assumptions in the next section to obtain a fully generalized multi-dimensional algorithm.
